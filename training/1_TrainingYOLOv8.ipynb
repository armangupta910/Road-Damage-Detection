```python
from ultralytics import YOLO
```

```python
# Enter training folder
%cd /home/armangupta/project/RoadDamageDetection/training
```

```python
# Project Name
_project = "runs/RDD_India"
_name = "Baseline_YOLOv8Small_Filtered"
_data = "/home/armangupta/project/RoadDamageDetection/training/dataset/rddIndiaFiltered/rdd_India.yaml"

# Epoch and warmup epoch
_epoch = 100
_warmup_epochs = int(_epoch*0.05)

model = YOLO('yolov8s.pt')
results = model.train(data=_data,
                      epochs=_epoch,
                      warmup_epochs=_warmup_epochs,
                      batch=32,
                      imgsz=640,
                      save_period=10,
                      workers=1,
                      project=_project,
                      name=_name,
                      seed=1337,
                      cos_lr=True,
                      mosaic=0.0,
                      
)
```

```python
# Resume Training
weight_path = "training/RDD_India/Baseline_YOLOv8Small_Filtered2/weights/last.pt"

model_Trained = YOLO(weight_path) # pass any model type
results = model_Trained.train(
    resume=True
)
```
